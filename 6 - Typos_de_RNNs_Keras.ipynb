## One-to-One
- No se usa el tiempo
- es equivalente a usar MLP

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, SimpleRNN
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(len(seq), 1, 1)
y = seq.reshape(len(seq), 1)
# define LSTM configuration
print('X'+str(X.shape)+':',X)
print('y'+str(y.shape)+':',y)
n_neurons = 2
# create LSTM
model = Sequential()
model.add(SimpleRNN(n_neurons, input_shape=(1, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
# train LSTM
n_batch = length
n_epoch = 2000
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=1)
```

```python
# evaluate
print('5 samples a la entrada', X.shape)
result = model.predict(X, batch_size=n_batch, verbose=0)
print('5 samples a la salida', result.shape)
for value in result:
    print('%.1f' % value)
```

Son 5 muestras, no es una secuencia ya que T=1, cada muestra de 1 feature

## Many-to-one

- Otro ejemplo sera el de generacin de texto como el de "Bodas de Sangre"
- Sentiment analisys
- Deteccin de SPAM
- Interesante utilizar Global Max pooling en vez de return_sequences en false. Para secuencias largas esto puede ser mas eficiente debido a la naturaleza de las RNN de quizas olvidar el largo plazo. (Ejemplo: Mail largo en deteccin de SPAM donde al ppio hay algo que determina muy claramente que es SPAM)

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length)
# define LSTM configuration
print(X.shape)
print(y.shape)
n_neurons = 2
# create LSTM
model = Sequential()
model.add(SimpleRNN(n_neurons, input_shape=(length, 1)))
model.add(Dense(length))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
# train LSTM
n_batch = 1
n_epoch = 1000
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
# evaluate
print(X.shape)
result = model.predict(X, batch_size=n_batch, verbose=0)
print(result.shape)
for value in result[0,:]:
    print('%.1f' % value)
```

En la entrada: Una sola muestra, T=5, 1 feature. 

A la salida de la RNN, solo toma las 2 neuronas de la ultima etapa temporal y las combina hacia una capa densa de 5 neuronas. 

La cantidad de pesos de la ultima capas es:

2x5 + 5 = 15

## Many-to-Many
- POS es un buen ejemplo
- Igual que tramas de video

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import TimeDistributed
from keras.layers import LSTM
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length, 1)
# define LSTM configuration
n_neurons = 2
# create LSTM
model = Sequential()
model.add(SimpleRNN(n_neurons, input_shape=(length, 1), return_sequences=True))
# Probar de cambiar Dense(2), etc
model.add(TimeDistributed(Dense(1)))
#model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
# train LSTM
n_batch = 1
n_epoch = 2000
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
print(X.shape)
print(result.shape)
for value in result[:,0]:
    print('%.1f' % value)
```

```python
result
```

La salida toma las 5 salidas temporales. Una neurona por cada T. Los pesos se comparten
